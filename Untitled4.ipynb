```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
def clean(data):
    data = data.drop(["campaign_id",  "iremoteid","device"], axis=1)
    return data

data = clean(data)
test = clean(test)
```

```python
def datetransform(x):
    x['start_date'] = pd.to_datetime(x['start_date'], format='%Y-%m-%d')
    x['start_date'] = x['start_date'].dt.year
    
    x['end_date'] = pd.to_datetime(x['end_date'], format='%Y-%m-%d')
    x['end_date'] = x['end_date'].dt.year

datetransform(data)
datetransform(test)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
columns = ["format"]

for col in columns:
    data[col] = le.fit_transform(data[col])
    test[col] = le.transform(test[col])
    print(le.classes_)
      
data.head(5)
```

```python
formats=pd.get_dummies(data['format'])
data = pd.concat([data,formats],axis=1)

data.head()
```

```python
formats=pd.get_dummies(test['format'])
test = pd.concat([test,formats],axis=1)
```

```python
data = data.drop(["format"], axis=1)
test = test.drop(["format"], axis=1)
```
